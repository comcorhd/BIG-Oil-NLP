```python
from stanza.utils.conll import CoNLL
import stanza
import os
import sys
sys.path.append("/home/elvis/ACDC-UD")
import estrutura_ud
```

```python
#stanza.download('pt')
```

```python
nlp = stanza.Pipeline('pt')
```

```python
arquivos = {}
arquivos_cru = {}
diretorio = "../50R/raw/Luisa"
for arquivo in os.listdir(diretorio):
    with open(f"{diretorio}/{arquivo}", encoding="utf-8") as f:
        text = f.read()
        arquivos[arquivo.rsplit(".txt")[0]] = CoNLL.convert_dict(nlp(text).to_dict())
        print(arquivo)
        print(nlp(text).entities)
        print("")
```

```python
sentences = []
conllus = {}
for arquivo in sorted(arquivos):
    conllus[arquivo] = []
    for s, sentence in enumerate(arquivos[arquivo]):
        text = " ".join([token[1] for token in sentence if not '-' in token[0]])
        sent_id = f"{arquivo}-{s+1}"
        sentences.append(f"# sent_id = {sent_id}\n# text = {text}\n" + "\n".join(["\t".join(token) for token in sentence]))
        conllus[arquivo].append(f"# sent_id = {sent_id}\n# text = {text}\n" + "\n".join(["\t".join(token) for token in sentence]))
```

```python
print("\n\n".join(sentences))
```

```python
with open("5R_stanza.conllu", "w", encoding="utf-8") as f:
    f.write("\n\n".join(sentences))
```

```python
resumos_golden = estrutura_ud.Corpus()
resumos_stanza = estrutura_ud.Corpus()
resumos_golden.load("../5R/5R_golden.conllu")
resumos_stanza.load("5R_stanza.conllu")

for sentid, sentence in resumos_golden.sentences.items():
    if sentid in resumos_stanza.sentences:
        for t, token in enumerate(sentence.tokens):
            if len(resumos_stanza.sentences[sentid].tokens) > t:
                if resumos_stanza.sentences[sentid].tokens[t].word == token.word:
                    resumos_stanza.sentences[sentid].tokens[t].misc = token.misc
resumos_stanza.save("5R_stanza_modified_misc.conllu")
```
